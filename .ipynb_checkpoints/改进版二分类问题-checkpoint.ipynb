```python
# 对于离散的字符串, 有两种处理方式
# 1. 转化成数字 2. 进行one-hot编码
data = data.join(pd.get_dummies(data.part)).join(pd.get_dummies(data.salary))
```

```python
# 把part和salary删掉
data.drop(columns=['part', 'salary'], inplace=True)
```
