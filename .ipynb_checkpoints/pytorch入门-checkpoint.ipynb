### 4张量的自动微分

```python
# 说白了就是自动求导
# 在要求导的变量定义中加上requires_grad=True表示要追踪这个变量的导数
x = torch.ones(2, 2, requires_grad=True)
y = x + 2
```

```python
x.requires_grad
```

```python
x2 = torch.ones(2, 2)
```

```python
# 默认是不会自动求导
x2.requires_grad
```

```python
y.grad_fn
```

```python
# 导数必须由标量输出创建, 意思就是说你要求的最后的结果, 必须是一个标量, 不能是向量
y.backward()
```

```python
# z = y.mean()
xy = y.sum()
```

```python
# 只有执行完backward之后, 所有requires_grad=True的变量的导数都会被自动求出
# z.backward()
xy.backward()
```

```python
# x.grad
x.grad
```

```python
z = y * y * 3
out = z.mean()
```

```python
out
```

```python
out.backward()
```

```python
y.grad
```

```python
# 求非叶子节点的tensor的导数, 必须使用retain_grad, 不然会报警告
y.retain_grad
```

```python
y.grad
```

```python
# 如果不需要求导, 可以把代码包在torch.no_grad()中
x.requires_grad
```

```python
(x ** 2).requires_grad
```

```python
# 通过with torch.no_grad() 暂时不对x求导
with torch.no_grad():
    print((x ** 2).requires_grad)
```

```python
x.requires_grad
```

```python
y = x ** 2 + 2
y.requires_grad
```

```python
y.detach()
```

```python
y.requires_grad
```

```python
x.detach()
```

```python
y = x ** 2 + 2
```

```python
y.requires_grad
```

```python
y = x.detach()
```

```python
y.requires_grad
```

```python
# 除了在定义变量的时候指定requires_grad, 也可以通过requires_grad_()方法修改变量的requires_grad属性
a = torch.randn(2, 2)
a = a * 3 + 2
print(a.requires_grad)
```

```python
a.requires_grad_(True)
```

```python
print(a.requires_grad)
```
