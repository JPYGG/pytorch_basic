```python
import torch
import numpy as np

# 打印cuda设备, 显卡名字
print(torch.cuda.get_device_name(0))
```

 ### 1 pytorch张量
 

```python
# pytorch中的张量和tensorflow的tensor是一样, 名字都一样
# pytorch中的张量也叫tensor
# tensor和numpy中的ndarray也是一个意思, 只不过tensor可以在GPU上加速计算
```

```python
# 创建tensor
torch.tensor([6, 2], dtype=torch.int32)
```

```python
torch.tensor((6, 2))
```

```python
torch.tensor(np.array([6, 2]))
```

```python
# 快速创建tensor的方法, 和numpy中的routines方法一样
# ones, zeros, full, eye, random.randn, random.normal...arange...  random.rand, random.random
# 创建一个0到1之间的随机数组成的tensor
# torch.rand((2, 3))
torch.rand(2, 3)
```

```python
# 标准正态分布
```

```python
torch.zeros(2, 3)
```

```python
torch.ones(2, 3)
```

```python
# tensor的shape
x = torch.ones(2, 3, 4)
x.shape
```

```python
# 通过x.size()方法获取形状
x.size()
```

```python
# size中可以传shape的索引
x.size(0)
```

### 2 tensor中的基础数据类型

- 32位浮点数: torch.float32
- 64位浮点数: torch.float64
- 32位正数: torch.int32
- 16位正数: torch.int16
- 64位正数: torch.int64

```python
# 我们可以在创建tensor时候指定数据类型
torch.tensor([6, 2], dtype=torch.float32)
```
